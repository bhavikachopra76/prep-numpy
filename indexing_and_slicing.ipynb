```python
import numpy as np
```

```python
#Indexing -> O-based indexing
#Accessing elements from array
a1 = np.array([1,2,3,4,5])
print(a1[0])
```

```python
#Negative Indexing
print(a1[-1])
```

```python
#Slicing [start:stop:step]
a2 = np.array([10,20,30,40,50])
print(a2[0:])
print(a2[0::2])
```

```python
#Reversing a list
print(a2[::-1])
```

```python
#Fancy Indexing -> Selecting multiple elements at a time (at different positions)
print(a2[[0,1,3,2]])
```

```python
#Boolean Masking -> Filtering elements of array
print(a2[a2>20])
```
