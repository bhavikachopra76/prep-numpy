```python
import numpy as np
```

```python
#Sorting
a1 = np.array([1,5,3,8,0])
a2 = np.sort(a1)
print(a2)
```

```python
#Sorting returning indexed
a3 = np.argsort(a1)
print(a3) #this returns indexed of the values in sorted order
```

```python
#Returning index of minimum and maximum value
print(np.argmax(a1))
print(np.argmin(a1))
```

```python
#Searching
print(np.where(a1>0)) #returns indexes
```
