```python
import numpy as np
```

```python
#Broadcasting -> Allows operations between arrays of different shapes by automatically expanding dimensions when possible.
a1 = np.array([[1,2,3], [4,5,6]]) #2d array
a2 = np.array([10,20,30])
result = a1 * a2
print(result)
```

```python
#Expanding single elements (Scalar vs Array Operations)
mul = a1 * 2
print(mul)
```

```python
#Does not happen between incompatible shapes
a3 = np.array([1,2])
# add = a1+a3
# print(add) this would give error
```

```python
#Vectorization -> performing operations on entire arrays at once instead of using Python loops.
addition = a1 + 5
print(addition)
```
