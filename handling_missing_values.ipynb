```python
import numpy as np
```

```python
#finding missing values
a1 = np.array([1,2,3,np.nan, 4,6, np.nan])
print(np.isnan(a1))
```

```python
#changing the missing values
cleaned_arr = np.nan_to_num(a1, nan = 10)
print(cleaned_arr)
print(np.isnan(cleaned_arr))
```

```python
#finding infinity values
a2 = np.array([1,2,3, np.inf, 5,6, -np.inf])
print(np.isinf(a2))
```

```python
#changing infinity values
cleared_arr2 = np.nan_to_num(a2, posinf=100 , neginf=-100)
print(cleared_arr2)
```
