```python
#colums : [Age, Math Marks, Science Marks]
data = np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
```

```python
#Get the shape of matrix
data.shape
```

```python
#Find the average age of students
np.mean(data[:,0])
```

```python
#Extract math marks of all students
data[:,1]
```

```python
#Find the highest science marks
np.max(data[:,2])
```

```python
#Get details of the students who scored more than 90 marks
data[data[:,1]>90]
```

```python
#Increase Math marks of all students by 5
data[:,1] + 5
```

```python
#Find how many students are younger than 19
len(data[data[:,0]<19])
```

```python
#Calculate the average marks in each subject(column - wise mean)
np.mean(data[:,1:] , axis = 0)
```
